```python
 #Now let create larger portfolio
all_vals = [boa['Adj. Close'], wfc['Adj. Close']]

stock_07_17 = pd.concat(all_vals, axis = 1)
stock_07_17.columns = ['boa Adj Close', 'wfc Adj Close']
```

```python
stock_07_17.head()
```

### Transform closimg price into instantaneous rates of returns

```python
returns = pd.DataFrame()
for stock in stock_07_17:
    if stock not in returns:
        returns[stock] =np.log(stock_07_17[stock]).diff()
returns= returns[1:]
returns.head()
```

### Generate some descritive ststistics

```python
returns.describe().transpose()
```

### Create Correlation matrix

```python
returns.corr()
```

### Graphically Display Correlation

```python
from pandas.plotting import scatter_matrix
scatter_matrix(returns,figsize=(16,12), alpha=0.3)
plt.show()
```
